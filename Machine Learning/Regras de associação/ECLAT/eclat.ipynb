```python
import pandas as pd

from pyECLAT import ECLAT
```

```python
# Carregar os dados
df = pd.read_csv("itemlist.txt", header = None)
df
```

```python
# Aplicar o algoritmo ECLAT
eclat = ECLAT(data = df)
eclat.df_bin
```

```python
# Exibindo valores das colunas
eclat.uniq_
```

```python
indices, support = eclat.fit(min_support = 0.02, min_combination = 1, max_combination = 3)
```
