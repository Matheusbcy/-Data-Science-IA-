## PyOD

```python
detector = KNN()
detector.fit(base_cov.iloc[:,0:12])
```

```python
prev = detector.labels_
np.unique(prev, return_counts = True)
```

```python
confiance = detector.decision_scores_
confiance
```

```python
outliers = []
for i in range(len(prev)):
  if prev[i] == 1:
    outliers.append(i)

list_outliers = base_cov.iloc[outliers, :]
list_outliers
```
