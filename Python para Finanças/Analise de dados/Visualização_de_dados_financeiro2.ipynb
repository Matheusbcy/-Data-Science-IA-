```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import pandas_datareader.data as web
```

```python
import yfinance as yf
```

```python
# Este código cria um DataFrame com os preços de fechamento das ações selecionadas a partir de 2015 usando o Yahoo Finance.
acoes = ["ABEV3.SA", "ODPV3.SA", "VIVT3.SA", "PETR4.SA", "BBAS3.SA", "BOVA11.SA"]
acoes_df = pd.DataFrame()

for acao in acoes:
    acoes_df[acao] = yf.download(acao, start="2015-01-01")["Close"]
```

```python
# Este comando exibe o DataFrame contendo os preços de fechamento das ações baixadas anteriormente.
acoes_df
```

```python
# Renomeia as colunas do DataFrame para nomes mais curtos e legíveis, e em seguida exibe o DataFrame atualizado.
acoes_df = acoes_df.rename(columns = {"ABEV3.SA": "AMBEV", "ODPV3.SA": "ODPV", "VIVT3.SA": "VIVO", "PETR4.SA": "PETRO", "BBAS3.SA": "BBRA", "BOVA11.SA": "BOVA"})
acoes_df
```

```python
# Verifica e exibe a quantidade de valores ausentes (nulos) em cada coluna do DataFrame.
acoes_df.isnull().sum()
```

```python
# Remove todas as linhas do DataFrame que contêm valores ausentes, garantindo que apenas dados completos sejam mantidos.
acoes_df.dropna(inplace = True)
```

```python
# Verifica novamente a presença de valores ausentes após a remoção das linhas com dados nulos.
acoes_df.isnull().sum()
```

```python
# Salva o DataFrame com os preços das ações em um arquivo CSV chamado "acoes.csv".
acoes_df.to_csv("acoes.csv")
```

```python
# Lê o arquivo CSV "acoes.csv" para um DataFrame e exibe seu conteúdo.
acoes_df = pd.read_csv("acoes.csv")
acoes_df
```

```python
# Exibe os nomes das colunas presentes no DataFrame carregado do arquivo CSV.
acoes_df.columns
```

```python
# Gera estatísticas descritivas (como média, desvio padrão, mínimos e máximos) para as colunas numéricas do DataFrame.
acoes_df.describe()
```

```python
# Cria um gráfico de linha interativo para visualizar o histórico dos preços das ações, plotando cada ação ao longo do tempo.
figure = px.line(title = "Histórico do preço das ações")
for i in acoes_df.columns[1:]:
  figure.add_scatter(x = acoes_df["Date"], y = acoes_df[i], name = i)
figure.show()
```

```python
# Normaliza os preços das ações dividindo cada valor pelo preço inicial, facilitando a comparação relativa da evolução das ações ao longo do tempo.
acoes_df_normalizado = acoes_df.copy()
for i in acoes_df_normalizado.columns[1: ]:
  acoes_df_normalizado[i] = acoes_df_normalizado[i] / acoes_df_normalizado[i][0]
```

```python
# Exibe o DataFrame com os preços das ações normalizados em relação ao valor inicial de cada ação.
acoes_df_normalizado
```

```python
# Plota um gráfico de linha interativo para visualizar a evolução normalizada dos preços das ações ao longo do tempo.
figure = px.line(title = "Histórico do preço das ações")
for i in acoes_df_normalizado.columns[1:]:
  figure.add_scatter(x = acoes_df_normalizado["Date"], y = acoes_df_normalizado[i], name = i)
figure.show()
```
